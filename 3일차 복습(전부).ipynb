### 1000이 넘는 43의 배수 중 가장 작은 수를 찾고 싶습니다.

```python
n = 0
while True:
    n = n + 1
    if n * 43 >= 1000:
        print(n * 43)
        break
```

```python
def jegop(x, y):
    a = 0
    b = 0
    for i in range(x, y + 1):
        a = a + i
        b = b + i**2
    return a**2 - b
```
