### 다이어트 관리 프로그램
- 목표 몸무게를 입력합니다.
- 현재 몸무게를 입력합니다.
- 매 주 감량한 몸무게를 입력한 뒤, 현재 몸무게가 목표 몸무게보다 적어지면, 프로그램을 종료합니다.
- 출력 값은 "축하합니다!__주 걸렸습니다."형태입니다.


```python
# 내가짠코드
a = int(input("현재 몸무게: "))
b = int(input("목표 몸무게: "))
n = 0
while True:
    c = int(input("감량한 무게: "))
    n = n + 1
    a = a - c
    if a <= b:
        print(f"축하합니다!{n}주 걸렸습니다.")
        break
```

```python
# 강사님이 짠 코드
weight = int(input("현재 몸무게: "))
goal = int(input("목표 몸무게를 입력하세요: "))
n = 0

while weight > goal:
    lost_weight = int(input("감량한 몸무게를 적어주세요: "))
    n = n + 1
    weight = weight - lost_weight

print(f"축하합니다! {n}주 걸렸습니다.")
 
```

### 1에서 100사이의 숫자 중 3의 배수인 값들의 합을 구하시오.

## 1에서 100사이의 숫자 중 소수의 합을 구하시오.
- 정답:1060

```python
n = 0
for i in range(2, 101):
    nmj = []
    for w in range(2, i):
        nmj.append(i%w)
    if 0 not in nmj:
        n = n + i
print(n)
```

### 업다운 게임
- 컴퓨터는 1부터 100 사이의 임의의 자연수를 생각합니다.
- 플레이어는 1부터 100 사이의 임의의 자연수를 input으로 입력합니다.
- 컴퓨터가 생각한 수가 플레이어가 입력한 수보다 크다면 Up, 작다면 Down을 출력합니다.
- 플레이어가 입력한 수가 컴퓨터가 생각한 수와 일치하면 몇번만에 맞췄는지 출력하고 프로그램을 종료합니다.

```python
# 내가 짠거
import random
num = random.randint(1, 100)
n = 0
for i in range(100):
    n = n + 1
    num1 = int(input("자연수: "))
    if num == int(num1):
        print("생각한 수")
        print(f"축하합니다! {n}번 만에 맞췄습니다.")
        break
    elif num  > int(num1):
        print("Up")
    else:
        print("Down")
```

```python
# 강사님이 짠거
import random
num  = random.randint(1, 100)

while True:
    a = int(input("숫자입력: "))
    
    if num > a:
        print("up")
        n = n + 1
    elif num < a:
        print("down")
        n = n + 1
    else:
        print(f"축하합니다! {}번만에 정답을 맞췄습니다.")
        break
```
